# Exploratory Data Analysis (EDA) - Stroke Prediction Dataset

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Ensure output directory for plots exists
os.makedirs("../data/eda_plots", exist_ok=True)
```

## 2. Load Data

```python
try:
    df = pd.read_csv('../data/raw/healthcare-dataset-stroke-data.csv')
    print("Data loaded successfully.")
    print(df.head())
except FileNotFoundError:
    print("Error: ../data/raw/healthcare-dataset-stroke-data.csv not found.")
    exit(1)
```

## 3. Data Inspection

```python
print("\n--- Data Info ---")
df.info()
```

```python
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# Duplicates
df.duplicated().sum()
```

```python
df.describe()
```

```python
for column in ['age', 'hypertension', 'heart_disease', 'avg_glucose_level',	'bmi', 'stroke']:
    sns.boxplot(df[column])
    plt.show()
```

```python
# Target Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='stroke', data=df)
plt.title('Distribution of Stroke (Target Variable)')
plt.savefig('../data/eda_plots/target_distribution.png')
print("\nTarget distribution plot saved to ../data/eda_plots/target_distribution.png")
```

```python
# Correlation Matrix
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig('../data/eda_plots/correlation_matrix.png')
print("Correlation matrix plot saved to ../data/eda_plots/correlation_matrix.png")
```
